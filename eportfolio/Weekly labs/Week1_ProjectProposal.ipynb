```python
import pandas as pd

# Load your dataset
data = pd.read_csv("predictive_maintenance.csv")  # Replace with your actual file path

# Print column names
print("Columns in your dataset:")
print(data.columns.tolist())
```

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histogram of Air Temperature (correct column: 'Air temperature [K]')
plt.figure(figsize=(8, 4))
sns.histplot(data['Air temperature [K]'], bins=20, kde=True)
plt.title("Distribution of Air Temperature")
plt.xlabel("Air Temperature (Kelvin)")
plt.ylabel("Frequency")
plt.show()

# 2. Boxplot of Rotational Speed by Failure Type (correct columns: 'Rotational speed [rpm]' vs 'Failure Type')
plt.figure(figsize=(10, 5))
sns.boxplot(x='Failure Type', y='Rotational speed [rpm]', data=data)
plt.title("Rotational Speed by Failure Type")
plt.xticks(rotation=45)  # Rotate x-labels for readability
plt.show()
```
